```python
import pymongo
import sqlite3
```

```python
#连接
conn = sqlite3.connect('E:/学习/大三上/当代数据管理系统/book_lx.db')
cur = conn.cursor()

# 查看DB文件里有哪些表
cur.execute("select * from sqlite_master").fetchall()
```

```python
book_attribute = ['book_id','title', 'author', 'publisher', 'original_title', 'translator', 'pub_year', 'pages',
                  'price', 'current_unit', 'binding', 'isbn', 'author_intro', 'book_intro', 'content', 
                  'tags', 'picture']
```

```python
host = '127.0.0.1'
port = 27017
client = pymongo.MongoClient(host, port)

# 切换数据库
db = client.bookstore # 没有会自动创建
```

```python
# 获取集合，如果没有则自动创建，在python中用users_col指代
users_col = db.books
```

```python
# 获取数据库中所有的书
sql = 'select * from book'
cur.execute(sql)
books = cur.fetchall()
print(books[0])
cur.close()
```

```python
book_tmp = [
    dict(zip(book_attribute, list(eachBook))) for eachBook in books
]
users_col.insert_many(book_tmp)

# for eachBook in books:
#     value = dict(zip(book_attribute, list(eachBook)[1:]))
#     users_col.insert_one(value)
# print("OK")
```
